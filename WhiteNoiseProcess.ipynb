```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
a = 0.9
x = np.empty(T+1)
x[0] = 0

for t in range(T):
    x[t+1] = a*x[t] + np.random.randn()
    
plt.plot(x, label="correlated time series")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
a = [0, 0.8, 0.98]
x = np.empty(T+1)
x[0] = 0

for alpha in a:
    for t in range(T):
        x[t+1] = alpha*x[t] + np.random.randn()
    
    plt.plot(x, label="time series alpha = " + str(alpha))
    plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
a = 0.9
x = np.empty(T+1)
x[0] = 0
e = 0

for t in range(T):
    e = np.random.randn()
    x[t+1] = a*abs(x[t]) + e

plt.plot(x, label="Time series absolute")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
a = 0.9
x = np.empty(T+1)
x[0] = 0
e = 0

for t in range(T):
    e = np.random.randn()
    if x[t] < 0:
        x[t+1] = a*(-x[t]) + e
    else:
        x[t+1] = a*x[t] + e

plt.plot(x, label="Time series manual abs")
plt.legend()
plt.plot()
```
