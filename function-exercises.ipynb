```python
def new_abs(num):
    return -num if num < 0 else num
```

```python
def generate_rand_data(size, generator_type):
    e_values = []
    for i in range(size):
        e = generator_type(0,1)
        e_values.append(e)
    return e_values

data = generate_rand_data(100, np.random.uniform)
plt.plot(data)
plt.show()
```

```python
def factorial(num):
    total = 1
    for i in range(num):
        total *= (i+1)
    return total

factorial(10)
```

```python
from numpy.random import uniform

def binomial_rv(n, p):
    count = 0
    for i in range(n):
        value = uniform(0,1)
        if value < p:
            count += 1
    return count

count = binomial_rv(2000, 0.5)
count
```

```python
from numpy.random import uniform

def distribucion_binomial(x, n, p):
    coeficiente_binomial = factorial(n) / (factorial(x)*factorial(n-x))
    return coeficiente_binomial * p**x * (1-p)**(n-x)

x = 0
n = 46
p = 0.65

probabilidad_de_valores = []
for x in range(n+1):
    probabilidad_de_valores.append(distribucion_binomial(x, n, p))
    
plt.plot(probabilidad_de_valores, label=f"n={n} p={p}")
plt.legend()
plt.show()

x = 30
p = 0
probabilidad_de_65 = []
for p in range(100+1):
    probabilidad_de_65.append(distribucion_binomial(x,n,p/100.0))
    
plt.plot(probabilidad_de_65, label=f"n={n} x={x}")
plt.legend()
plt.show()
```

```python
from numpy.random import uniform

def consecutive_head(k):
    count_consecutive = 0
    for i in range(10):
        if uniform(0,1) < 0.5:
            count_consecutive += 1
        else:
            count_consecutive = 0
        if count_consecutive >= k:
            return 1
    return 0

def hit_k_head(k):
    total = 0
    for i in range(10):
        if uniform(0,1) < 0.5:
            total += 1
        if total >= k:
            return 1
    return 0
```
